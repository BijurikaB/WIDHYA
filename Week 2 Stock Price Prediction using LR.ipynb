```python
data = quandl.get("EOD/AAPL")
data
```

#### Meaningful features to get an insight into the future prices: High Low Percentage and Percentage Change

```python
#High low Percentage
data['HL_PCT']= (data['Adj_High'] - data['Adj_Low'])/data['Adj_Close']*100
#Percentage Change
data['PCT_change']= (data['Adj_Close'] - data['Adj_Open'])/data['Adj_Open']*100
```

```python
data
```

##### New dataframe with the new features

```python
new_data = data[['Adj_Close','Adj_Volume','HL_PCT','PCT_change']]
new_data.head()
```

```python
new_data.isna().sum()
```

No NaN values found

```python
out = 10 #days
new_data['Prediction'] = new_data[['Adj_Close']].shift(-out)
new_data
```

```python
new_data.tail()
```

```python
new_data.isna().sum()
```

There are NaN values

#### Conversion to numpy array 

```python
X = np.array(new_data.drop(['Prediction'],1))
X = X[:-out]
print(X)
```

```python
y = np.array(new_data['Prediction']) 
y = y[:-out]
print(y)
```

##### Split the dataset into 80% training and 20% testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
print('Training Score: ', model.score(X_train,y_train))

print('Testing Score: ', model.score(X_test,y_test))
```

```python
y_pred = model.predict(X_test)
y_pred
```

#### Accuracy of the model

```python
acc=metrics.r2_score(y_test,y_pred)
print("Accuracy Score of Model: ",round(acc*100,2),'%')
```

#### Actual Price and Predicted Price

```python
prediction_df = pd.DataFrame({ 'Actual Price': y_test, 'Predicted Price': y_pred})  
prediction_df
```

#### Model evaluation

```python
print('Mean Squared Error: ',mean_squared_error(y_test,y_pred))
print('Coefficient of determination: ',r2_score(y_test,y_pred))
print('Mean Squared Error: ',mean_absolute_error(y_test,y_pred))
```

#### Forecasting

```python
def predictor(forecast_days):
    
    
    # A variable for predicting 'n' days out into the future
    out = forecast_days
    new_data['Prediction'] = new_data[['Adj_Close']].shift(-out)
    
    # Convert the dataframe to a numpy array
    X = np.array(new_data.drop(['Prediction'],1))
    
    #Remove the last 'n' rows
    X = X[:-out]
    
    # Convert the dataframe to a numpy array 
    y = np.array(new_data['Prediction'])
    
    # Get all of the y values except the last '10' rows
    y = y[:-out]

    # Split the data into 80% training and 20% testing
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    
    # Create and train the Linear Regression  Model
    model = LinearRegression()
    
    # Train the model
    model.fit(X_train, y_train)

    # The best possible score is 1.0
    model_confidence = model.score(X_test, y_test)
    print('The Confidence {}\n'.format(model_confidence))
    
    # Set x_forecast equal to the last 10 rows of the original data set from Adj. Close column
    x_forecast = np.array(new_data.drop(['Prediction'],1))[-out:]

    
    # Print linear regression model predictions for the next '10' days
    model_prediction = model.predict(x_forecast)
    print('Prediction for next {} day price:\n {}'.format(forecast_days,model_prediction))
```

```python
predictor(1)
```
